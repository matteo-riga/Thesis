```python
import json
import pandas as pd
import os
```

```python
fp = existing_file_paths[0]
data_list = []
temp = []
start_line = 72160
current_line = 0
i = 0
    
with open(fp, 'r') as file:
    for line in file:
        current_line += 1
        if current_line < start_line:
            continue  # skip lines until reaching the starting line
        #print(f"line {current_line}")
        temp.append(line)
        line = line.strip()
        i += 1
    
        if line:
            try:
                data = json.loads(line)
                data_list.append(data)
            except json.JSONDecodeError as e:
                print(e)
                continue
```

```python
df = pd.DataFrame(data_list)
df.head()
```

```python
df.columns = ['hostname', 'ident', 'message', 'severity', 'facility', 'time']
```

```python
p = ParamsExtractor3.ParamsExtractor(df)
df = p.convert_params(df)
#df.columns = ['hostname', 'ident', 'message', 'severity', 'facility', 'time',
       #'severity_numbers', 'facility_numbers', 'severity_scores']
l = LaurelParamsExtractor.LaurelParamsExtractor()
new_df = l.get_params(df)
#df_after.append(new_df)
df = pd.concat([df, new_df], axis=1)
df.head()
```
