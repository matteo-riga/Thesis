```python
# First type of log file:
# laurel_anomalous_new
fp = '../data/linpeas_logs_2_silent_attack/laurel_logs.log'
fps = [fp]
fps += [fp + f'.{str(i)}' for i in range(1,8)]
print(fps)
```

```python
fp = fps[0]

data_list = []
lines_limit = 1000 # limit reading lines, done for memory constraints
line_from = 183662 - 1000
i = 0

with open(fp, 'r') as file:
    for line in file:
        i+=1
        if i > line_from and i < line_from + lines_limit:
            index = line.find('{')
            if index != -1:
                json_data = line[index:]
                data = json.loads(json_data)
                data_list.append(data)
                
df = pd.DataFrame(data_list)
```

```python
import re

def parse_audit_log(log: str):
    # Initialize an empty dictionary to hold the key-value pairs
    result = {}

    # Use regular expressions to find all key-value pairs
    # The pattern matches sequences of the form key=value or key="value"
    matches = re.findall(r'(\w+)=("[^"]+"|\S+)', log)

    # Iterate over all matches and add them to the dictionary
    for key, value in matches:
        # Remove quotes from the value if they exist
        result[key] = value.strip('"')

    # columns to keep: cwd/exe, exit, items, ppid, pid, comm, timedelta, pid timedelta, id_anomalies, num_id_anomalies
    
    
    return result

# Example usage
# log = 'type=SYSCALL msg=audit(1720451679.133:2225): arch=c000003e syscall=59 success=yes exit=0 a0=7f64b5abd8d8 a1=7fff22281de0 a2=7fff22282678 a3=0 items=2 ppid=3879 pid=3880 auid=1000 uid=0 gid=0 euid=0 suid=0 fsuid=0 egid=0 sgid=0 fsgid=0 tty=pts0 ses=3 comm="systemd-tty-ask" exe="/usr/bin/systemd-tty-ask-password-agent" subj=unconfined_u:unconfined_r:unconfined_t:s0-s0:c0.c1023 key="exec_log"\x1dARCH=x86_64 SYSCALL=execve AUID="matteo" UID="root" GID="root" EUID="root" SUID="root" FSUID="root" EGID="root" SGID="root" FSGID="root"'

parsed_log = parse_audit_log(df['msg'][0])
print(parsed_log)
```

```python
# Apply the parse_audit_log function to each row in the 'log' column
df_expanded = df['msg'].apply(parse_audit_log).apply(pd.Series)

# Concatenate the new columns with the original DataFrame
df = pd.concat([df, df_expanded], axis=1)
df.head()
```

```python
print(df.columns.values)
```

```python
cols_to_keep = ['timereported', 'exe', 'exit', 'items', 'ppid', 'pid', 'comm', 'auid', 'uid', 'gid', 'euid', 'suid', 'fsuid', 'egid', 'sgid', 'fsgid', 'AUID', 'UID', 'GID', 'EUID', 'SUID', 'FSUID', 'EGID', 'SGID', 'FSGID']
df = df[cols_to_keep]
df.head()
```

```python
df['timereported'] = pd.to_datetime(df['timereported'])
df['timestamp_numeric'] = df['timereported'].astype('int64') // 10**6
df['timedelta'] = df['timestamp_numeric'].diff()
df.head()
```

```python
df = df.drop(['timereported', 'timestamp_numeric'], axis=1)
df.head()
```

```python
for col in df.columns:
    print(f'{col} : {df.iloc[1][col]}')
```

```python
columns = ['auid', 'uid', 'gid', 'euid', 'suid', 'fsuid', 'egid', 'sgid', 'fsgid']

def check_row_id_num(row):
    return pd.Series({col: row[col] == 0 for col in columns})

checks = df.apply(check_row_id_num, axis=1)
df['num_id_anomalies'] = checks.sum(axis=1)
df = df.drop(columns, axis=1)
```
